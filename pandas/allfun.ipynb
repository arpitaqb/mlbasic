grouping functions

```python
import pandas as pd
import numpy as np
```

```python


data = {
    'department': ['Sales', 'Sales', 'HR', 'HR', 'IT'],
    'employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'salary': [50000, 60000, 45000, 47000, 70000]
}
df = pd.DataFrame(data)
df
```

```python
df.groupby('department')['salary'].std()
```

```python
df.groupby(['department', 'employee'])['salary'].mean()
```

```python
df.groupby('department')[['salary']].agg(['sum', 'mean'])
```

```python
df.groupby('department').agg(total_salary = ('salary', 'sum'),
                             average_salary = ('salary', 'mean'))
```

```python
df['department'].value_counts()
```

```python
df.pivot_table(index='department', values='salary', aggfunc='mean')
```
