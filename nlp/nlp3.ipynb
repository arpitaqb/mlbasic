### word2vec

```python
import nltk
```

```python
from gensim.models import Word2Vec
```

```python
sen = [
    ["i", "love", "deep", "learning"],
    ["deep", "learning", "is", "fun"],
    ["i", "enjoy", "machine", "learning"]
]
```

```python
model = Word2Vec(sen, vector_size=100, window=2, min_count=2, sg=1) #train word2vec model 
#sg = 1 uses skip-gram and if 0 then it is using cbow
```

```python
vector = model.wv['learning'] #vector for a word
vector
```

```python
model.wv.most_similar('learning') #find most similar words 
```

```python
model = Word2Vec(sen, vector_size=50, window=2, min_count=2, sg=1) 
model.wv['learning']
```

```python
model.wv.most_similar('learning')
```

### avgword2vec

```python
from gensim.models import Word2Vec
import numpy as np
```

```python
sen =  ["i", "love", "deep", "learning"]
```

```python
model = Word2Vec(sen, vector_size=100, window=2, min_count=1, sg=1)
```

```python
vector = [model.wv[w] for w in sen if w in model.wv]
```

```python
if vector:
    avg_vector = np.mean(vector, axis=0)
    print(avg_vector)
else:
    print("no words in sentence")
```
