```python
from numpy import random 
import numpy as np
```

```python
a = random.randint(20) #give random int number between given range
a
```

```python
b = random.random((3,4)) #give array of n elements
b
```

```python
c = random.rand() #give random float number of range 
c
```

```python
d = random.randint(40, size=(3,5)) # it has size parameter you can define any size of array you want 
d
```

```python
e = random.choice([1,5,3])
print(e)  #it will always give value from choice 

f = random.choice([4,8,2,5,3,9],size=(3,4))
f
```

```python
g = random.choice([1,2,3,4], p=[0.4, 0.3,0.3,0.0] , size=(50)) #no matter how many time it will run it never give 4 as it's probability is 0 
g
```

```python
#suffle element

h = np.array([1,2,3,4,5])

print("permutation", random.permutation(h)) #make new array do not change original array
print("original array",h)
random.shuffle(h)  #make changes in original array
print("suffle",h)
```

```python
np.random.seed(0) # it will fix the seed which no matter how many time code run output will be same 
a = np.random.random((3,4))
a
```
