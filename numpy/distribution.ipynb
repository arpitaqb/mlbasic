```python
#1.normal distribution : normal data distribution 
from numpy import random 

a = random.normal(size=(2,3))
a
```

loc = mean (top of the graph)

scale = standard deviation(how much flattern should be)

size = return shape of array

n = number of trials

p = probability of occurance 

lam = known number of occurance

low = lower bound

high = upper bound 

pvals = list of probabilities of outcomes

df = degree of freedom

a = shape parameter

```python
a = random.normal(loc=1, scale=2, size=(2,3))
a
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(size=100),kind="kde")
plt.show()
```

```python
sns.distplot(random.normal(size=1000), hist=False)
plt.show()
```

```python
#2. bionomial distribution : discrete distribution(describes the outcomes of binary scenarios)

b = random.binomial(n=10, p=0.5, size=10)
b
```

```python
sns.displot(random.binomial(n=10, p=0.5, size =100),kind="kde")
plt.show()
```

```python
data = {
  "normal": random.normal(loc=50, scale=5, size=1000),
  "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

```python
#3. poisson distribution : it is a discreate distribution.(how many times an event can happend in a specific time.)
c = random.poisson(lam=2, size=100)
c
```

```python
sns.displot(random.poisson(lam=2, size=100),kind="kde")
plt.show()
```

```python
data = {
    "normal" : random.normal(loc=10, scale=5, size=100),
    "bionomial": random.binomial(n=3, p=0.5, size=100),
    "poisson" : random.poisson(lam=3, size=100)
}

sns.displot(data, kind="kde")
plt.show()
```

```python
#4. uniform distribution : used to discribe probability where every event has qual chance of accuring.
d = random.uniform(low=0.2, high=0.9, size=(2,3))
d
```

```python
sns.displot(random.uniform(size=100),kind="kde")
plt.show()
```

```python
#5. logistic distribution : used to describe growth.
e = random.logistic(loc=1, scale=2, size=(2,3))
e
```

```python
sns.displot(random.logistic(size=100),kind='kde')
plt.show()
```

```python
#6. multinomial distribution : generalization on bionomial 
f = random.multinomial(n=4,pvals=[1/4,1/4,1/4,1/4])
f
```

```python
import numpy as np
g = np.random.multinomial(n=4, pvals=[1/3, 1/3, 1/3], size=1000)
sns.displot(g, kind="kde")
plt.show()
```

```python
#7.exponential distribution : used for describing time till next event
#here scale defines inverse of rate (default to 1)

h = random.exponential(scale=3, size=(3,4))
print(h)

sns.displot(random.exponential(size=100),kind="kde")
plt.show()
```

```python
#8.chi square distribution : used as basis to verify the hypothesis.

i = random.chisquare(df=2, size=100)
print(i)

sns.displot(i, kind="kde")
plt.show()
```

```python
#9.rayleigh distribution : used in single processing
j = random.rayleigh(scale=2, size=10)
print(j)

sns.displot(j, kind="kde")
plt.show()
```

```python
#10. pareto distribution : follows pareto's law:80-20 distribution
k = random.pareto(a=2, size=(2,3))
print(k)

sns.displot(random.pareto(a=2,size=100))
plt.show()
```

```python
#11. zipf distribution : based on zipf law(nth common term is 1/n times of most common term)
#here a is distribution parameter
l = random.zipf(a=2, size=(2,3))
print(l)

m = random.zipf(a=2, size=1000)
sns.displot(m[m<10])
plt.show()
```
